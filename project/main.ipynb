### Use kaggle to download dataset about French Presidential Election

In this example we will use kagglehub to download a dataset about the French Presidential Election from Kaggle.
Dataset URL: https://www.kaggle.com/datasets/jeanmidev/french-presidential-election/data

```python
!pip install kagglehub
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jeanmidev/french-presidential-election")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os
import sqlite3

#os.remove("/home/bl4ck/.cache/kagglehub/datasets/jeanmidev/french-presidential-election/versions/21/database_googletrends.sqlite")

dfs_list = []
mention='"mention_Le PEN"'

for file in os.listdir(path):
    print("File in dataset:", file)

    file_path = os.path.join(path, file)
    print("Loading file:", file_path)

    connection = sqlite3.connect(file_path)
    print("Reading the database")
    sql_query = f"SELECT text FROM data WHERE lang='en' AND "+mention+"=1;"
    dfs_list.append(pd.read_sql_query(sql_query, connection))
    connection.close()

#concat tweets of all database on one list
dfs_list = pd.concat(dfs_list)
print("Number of tweets mentioning Le PEN in English:", len(dfs_list))
print(dfs_list.head(10))
```
